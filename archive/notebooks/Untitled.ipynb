```python
import json
from tqdm import tqdm
from matplotlib import pyplot as plt
```

```python
with open("./collection.json", "r") as file_handle:
    notion_docs = json.load(file_handle)
```

```python
doc_lens = []

for doc in tqdm(notion_docs["documents"]):
    
    doc_len = len(doc)
    doc_lens.append(doc_len)
```

```python
a, b, _ = plt.hist(doc_lens, bins = 1000)
```
